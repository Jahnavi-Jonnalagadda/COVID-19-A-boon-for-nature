```python
import matplotlib.pyplot as plt
import statsmodels
from statsmodels.tsa.stattools import adfuller
import pandas as pd
```

```python
# ADF test to check the stationarity od series
def perform_ADF(series):
    # Akaike Information Criterion (AIC) is used to determine the lag
    adfTest = adfuller(series, autolag='AIC')
    print('ADF Statistic:',adfTest[0])
    print('n_lags:',adfTest[1])
    print('p-value:',adfTest[1])
    for key, value in adfTest[4].items():
        print('Critial Values:')
        print(key,value) 
        
    p_value = adfTest[1]
    critical_value = adfTest[4]['5%']
    # compare the p_value with 5% critical value
    # if p_value > critcal value it is non stationary
    # if p_value <= critical_value it is stationary
    if p_value <= critical_value:
        print('Can Reject Null Hypothesis')
        print('Series is Stationary')
    else:
        print('Can not Reject Null Hypothesis')
        print('Series is Non-Stationary')     
```

```python
import random
random.seed(5)
```

```python
# series with a trend
time_series = []
for i in range(100):
    time_series.append(random.random() + (i*0.02))

plt.plot(time_series)
plt.show()
perform_ADF(time_series)
```
